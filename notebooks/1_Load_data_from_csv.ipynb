```python
import graphlab as gl
```

# Reading The CSV Files

```python
papersSF = gl.SFrame.read_csv('../data/csvData/abstract_data.csv', error_bad_lines=False,
                              delimiter = "\t",
                              na_values = ['N/A'],
                              column_type_hints = {'ID' : int,
                                                  'Year': int,
                                                  'Pages': int})
papersSF
```

```python
citationSF = gl.SFrame.read_csv('../data/csvData/hep-th-citations.csv',
                                delimiter = " ",
                                column_type_hints = {'SourceID' : int,
                                                     'TargetID' : int})
citationSF
```

# Creating The SGraph

```python
paperPaperSG = gl.SGraph()
paperPaperSG = paperPaperSG.add_vertices(vertices=papersSF, vid_field='ID')
paperPaperSG = paperPaperSG.add_edges(edges=citationSF, src_field='SourceID', dst_field='TargetID')
```

```python
paperPaperSG.get_edges().to_dataframe()
```

# Cleaning the pages 

```python
papersSF['Pages'].apply(lambda x : None if x > 1000 else x)
```

# Addding Bag of Words

```python
papersSF['Title_BOW'] = papersSF['Title'].count_words()
'''
    BOW: bag-of-words dictionary
    returns newBOW = bag 
'''
def delete_2_char_words(BOW):
    newBOW = {}
    for key,value in BOW.items():
        if len(key) > 2:
            newBOW[key] = value
    return newBOW

papersSF['Title_BOW'] = papersSF.apply(lambda x: delete_2_char_words(x['Title_BOW']))

papersSF['Title_BOW']
```
