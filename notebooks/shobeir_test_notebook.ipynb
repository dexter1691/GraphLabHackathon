# Loading the papers

```python
papersSF = gl.SFrame.read_csv('../data/csvData/papers.csv', error_bad_lines=True,
                             column_type_hints = {'PaperID' : int,
                                                  'Date': int,
                                                  'Pages': int})
papersSF
```

```python
citationSF = gl.SFrame.read_csv('../data/csvData/hep-th-citations.csv',
                                delimiter = " ",
                                column_type_hints = {'SourceID' : int,
                                                     'TargetID' : int})
citationSF[citationSF['SourceID'] == 9506048]
```

```python
#paperSF.save('../data/glData/papers_org')
```

# Creating the graph

```python
paperPaperSG = gl.SGraph()
paperPaperSG = paperPaperSG.add_vertices(vertices=papersSF, vid_field='PaperID')
paperPaperSG = paperPaperSG.add_edges(edges=citationSF, src_field='SourceID', dst_field='TargetID')
```
